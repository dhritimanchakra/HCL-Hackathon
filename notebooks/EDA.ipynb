```python
import pandas as pd

# 1. Read the dataset from your data folder
df = pd.read_csv('../data/train.csv')

# 2. Define our Expected Schema 
# (These are the critical columns we need for our custom Indian real estate logic)
expected_schema = {
    'MSZoning': 'object',    # We need this to map Chennai vs. Vellore
    'YrSold': 'int64',       # We need this for the 2026-2028 timeline
    'BldgType': 'object',    # We need this to identify apartments
    'PoolArea': 'int64',     # We need this for amenities
    'OverallQual': 'int64',  # We need this for the gym logic
    'SalePrice': 'int64'     # Our target variable
}

# 3. Perform Schema Validation
def validate_dataset(data, schema):
    print("--- Starting Schema Validation ---")
    
    # Check 1: Do all necessary columns exist?
    missing_cols = [col for col in schema.keys() if col not in data.columns]
    if missing_cols:
        raise ValueError(f"❌ Validation Failed! Missing critical columns: {missing_cols}")
    else:
        print("✓ All required columns are present.")
        
    # Check 2: Are the data types correct?
    type_errors = False
    for col, expected_type in schema.items():
        actual_type = str(data[col].dtype)
        if expected_type not in actual_type:
            print(f"⚠️ Warning: '{col}' is type {actual_type}, but expected {expected_type}")
            type_errors = True
            
    if not type_errors:
        print("✓ All data types match the expected schema.")
        
    print("--- Schema Validation Passed! Ready for Preprocessing ---")

# Run the validation
validate_dataset(df, expected_schema)

# Print the shape to confirm loading
print(f"\nDataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} columns.")
```
